```python
def make_double(img):
    
    h = img.shape[0]
    w = img.shape[1]
    
    img2 = []
    f = 0;
    
    for i in range(0, h):
        arr = []
        zz = []
        for j in range(0, w):
                arr.append(img[i][j])
                arr.append(0)
                zz.append(0)
                zz.append(0)
        img2.append(zz)
        img2.append(arr)
    
    img3 = []
        
    for i in range(0, 2*h):
        arrr = []
        for j in range(0, 2*w):
            if( img2[i][j] == 0 ):
                if( i-1 >= 0 and img2[i-1][j] != 0):
                    arrr.append(img2[i-1][j])
                elif( j-1 >=0 and img2[i][j-1] != 0):
                    arrr.append(img2[i][j-1])
                elif( i-1 >= 0 and j-1 >=0 ):
                    arrr.append(img2[i-1][j-1])
                else:
                    arrr.append(img2[i][j])
            else:
                arrr.append(img2[i][j])
        img3.append(arrr)
        
        plt.imshow(img3, cmap="gray")
        return img3
    
```
