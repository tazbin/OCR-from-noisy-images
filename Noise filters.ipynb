```python
import cv2
from matplotlib import pyplot as plt
from skimage.util import random_noise
```

```python
I = cv2.imread('images/bb.jpg', 1)
gauss = random_noise(I, mode='gaussian', seed=None, clip=True)
sp = random_noise(I, mode='s&p', seed=None, clip=True)
localvar = random_noise(I, mode='localvar', seed=None, clip=True)
passion = random_noise(I, mode='poisson', seed=None, clip=True)
salt = random_noise(I, mode='salt', seed=None, clip=True)
pepper = random_noise(I, mode='pepper', seed=None, clip=True)
speckle = random_noise(I, mode='speckle', seed=None, clip=True)
```

```python
plt.imshow(I), plt.title('Origin')
```

```python
plt.imshow(gauss), plt.title('Gaussian')
```

```python
plt.imshow(sp), plt.title('Salt & pepper')
```

```python
plt.imshow(localvar), plt.title('Localvar')
```

```python
plt.imshow(passion), plt.title('Poisson')
```

```python
plt.imshow(salt), plt.title('Salt')
```

```python
plt.imshow(speckle), plt.title('Speckle')
```

```python
for i in range(0, 10):
    I = random_noise(I, mode='salt', seed=None, clip=True)
    I = random_noise(I, mode='pepper', seed=None, clip=True)

plt.imshow(I), plt.title('Merged')
I = cv2.imread('images/bb.jpg', 1)
```
